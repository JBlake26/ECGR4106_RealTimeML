# Problem 1

```python
imgblue = Image.open("/content/drive/MyDrive/ECGR 4106 - Real Time Machine Learning/Colab Notebooks/HW1/blue.jfif")
imgblue
```

```python
transform = transforms.ToTensor()
imgblueTensor = transform(imgblue)
print("Shape of Blue Image Tensor: \n", imgblueTensor.shape)
```

```python
R_mean, G_mean, B_mean = torch.mean(imgblueTensor, dim = [1,2])
print("Mean across Red channel:", R_mean)
print("Mean across Green channel:", G_mean)
print("Mean across Blue channel:", B_mean)
```
